```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

# Correlation with heatmap

###### A correlation heatmap uses the color scale, To show a 2D correlation relation matrix(table). between 2 descrite dimensions or event type. 
###### It is very import in feature selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Jointplot

###### A jointplot allows you to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatter plot, a hexbin plot,a 2D histogram or a 2D desisty plto. 

# Univariate analysis

```python
sns.jointplot(x='tip', y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair plot

##### A pair plot is also known as scatter plot, in which one variable in the same data row is matched with another variable's value like this.

```python
sns.pairplot(df)
```

```python
sns.pairplot(data=df,hue='sex')
```

```python
sns.pairplot(data=df,hue='smoker')
```

```python
sns.pairplot(data=df,hue='day')
```

```python
sns.pairplot(data=df,hue='time')
```

# Distplot

###### Distplot helps you to create the histogram. Dist plot helps us to check the columns feature.If kde is present it will show the density.if not it will show the count.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical plot

###### Seaborn also helps us in doing the analysis on the Categorical data.

###### Countplot
###### barplot
###### boxplot
###### violinplot

```python
## count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot('time',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
##Bar plot
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.barplot(x='smoker',y='tip',data=df)
```

```python
sns.barplot(x='day',y='tip',data=df)
```

# Box plot

###### Box plot or whisker plot is a graph that presents information from a five number summary

```python
sns.boxplot('smoker','tip',data=df)
```

```python
sns.boxplot('sex','tip', data=df)
```

```python
sns.boxplot('day','tip',data=df)
```

```python
sns.boxplot('time','tip',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot('sex','tip',hue='smoker',data=df,)
```

```python
sns.boxplot('day','tip',hue='smoker',data=df)
```

# Violin plot

##### violinplot helps us to see both the distribution of data in terms of kernal density estimation and the box plot

```python
sns.violinplot('total_bill','day',data=df)
```
