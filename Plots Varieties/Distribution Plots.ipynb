### We can distribution plots with histogram

```python
import pandas as pd
import numpy as np
pstore=pd.read_csv("googleplaystore.csv")
pstore.head(10)

# Importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(pstore.Rating)
plt.show()
```

```python
# Change the number of bins
sns.histplot(pstore.Rating,bins=20,kde=False)
plt.show()
```

```python
# Importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a distribution plot for rating
sns.distplot(pstore.Rating,bins=20,color="g")
plt.title("Distribution of app ratings", fontsize=20,color='red')
plt.show()
```

### Scatter plots

```python
import seaborn as sns
tips=sns.load_dataset("tips")
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
sns.scatterplot(data=tips,x="total_bill", y="tip",hue="time")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",style="time")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time")
```

### Heat maps

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
sns.set()
uniform_data=np.random.rand(10,12)
ax=sns.heatmap(uniform_data,vmin=0,vmax=1)
plt.show()
```

### Heatmap Colors

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(np.random.randn(50).reshape(10,5))
corr=df.corr()
```

```python
ax1=sns.heatmap(corr,cbar=0,linewidths=2,vmax=1,vmin=0,square=True,cmap='Blues')
plt.show()
```
