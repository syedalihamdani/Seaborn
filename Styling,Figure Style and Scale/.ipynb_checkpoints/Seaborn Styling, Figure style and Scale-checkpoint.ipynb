### Built in Themes

```python
get_ipython().run_line_magic('matplotlib','inline')
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

### To use any kind of theme sns.set_style()

```python
# By default Seaborn use the darkgrid theme
sns.set_style("darkgrid")
sns.stripplot(x="day",y="total_bill",data=tips)
```

### Background Color 

```python
sns.set_style("dark")
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
#  White and ticks makes color more visible for your data
sns.set_style("ticks")
sns.stripplot(x="day",y="total_bill",data=tips)
```

### Grids

```python
sns.set_style("whitegrid")
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.set_style("white")
sns.stripplot(x="day",y="total_bill",data=tips)
```

### Despine

```python
sns.set_style("white")
sns.stripplot(x="day",y="total_bill",data=tips)
sns.despine()
```

```python
sns.set_style("whitegrid")
sns.stripplot(x="day",y="total_bill",data=tips)
sns.despine(left=True,bottom=True)
```

### Scaling Figure Styles for Different Mediums

#### Scaling Plots

```python
sns.set_style("ticks")
# smallest context
```

```python
sns.set_context("paper")
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.set_style("ticks")
# Largest Context
sns.set_context("poster")
sns.stripplot(x="day",y="total_bill",data=tips)
```

### Scaling Fonts and Line widths

```python
# Set font scale and reduce grid line width to match
sns.set_context("poster",font_scale=.5,rc={"grid.linewidth":0.6})
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
# Set font scale and increase grid line width to match
sns.set_context("poster",font_scale=1,rc={"grid.linewidth":5})
sns.stripplot(x="day",y="total_bill",data=tips)
```

### RC Parameter

```python
sns.set_style("ticks")
sns.set_context("poster")
sns.stripplot(x="day",y="total_bill",data=tips)
sns.plotting_context()
```

### Temporarily Setting The Plot Style

### Import necessary libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
```

```python
iris=sns.load_dataset("iris")
tips=sns.load_dataset("tips")
```

```python
# Set axes style to white for first subplot
```

```python
with sns.axes_style("white"):
    plt.subplot(211)
    sns.swarmplot(x="species",y="petal_length",data=iris)
    # Initialize second subplot
    plt.subplot(212)
    sns.violinplot(x="total_bill",data=tips)
#     Show the plot
plt.show()
```

### Setting the figure size

```python
# Import necessary libraries
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Initalize Figure and Axes object
fig,ax=plt.subplots(figsize=(10,4))

# Load in the data
iris=sns.load_dataset("iris")

# Create swarmplot
sns.swarmplot(x="species",y="petal_length",data=iris,ax=ax)

# Show plot
plt.show()
```
